#### Problem 2 - Load the Snares and Kicks tracks and extract features.

Be sure to extract at least four features, which might include [Energy and RMSE](https://musicinformationretrieval.com/energy.html), [Zero Crossing Rate](https://musicinformationretrieval.com/zcr.html) and/or any of the [spectral features](https://musicinformationretrieval.com/spectral_features.html)

```python
import glob
import librosa
```

```python
kicks = [librosa.load(x)[0] for x in glob.glob("./Dubstep Empire Drum Kit/Kicks/*.wav")]
snares = [librosa.load(x)[0] for x in glob.glob("./Dubstep Empire Drum Kit/Snares/*.wav")]
combined_drums = kicks + snares
```

```python
len(kicks), len(snares), len(combined_drums)
```

```python
# https://librosa.org/doc/main/feature.html

def extract_features(signal):
    return (
        librosa.feature.zero_crossing_rate(signal)[0, 0],
        librosa.feature.spectral_centroid(signal)[0, 0],
        librosa.feature.spectral_rolloff(signal)[0,0],
        librosa.feature.tonnetz(signal)[0,0]
    )
```

```python
features = [extract_features(s) for s in combined_drums]
```

```python
features[:5]
```

```python
import pandas as pd

combined_data = pd.DataFrame(features, columns = ["zcr", "spec_cent", "spec_rolloff", "tonnetz"])
combined_data["type"] = 503*["kick"] + 415*["snare"]
combined_data.head()
```

#### Problem 3 - Make plots of the feature space

Before building the classifier, you should make various plots of the feature space using [Seaborn pairplot](https://pythonbasics.org/seaborn-pairplot/)

```python
import seaborn as sb
sb.pairplot(combined_data, hue="type")
```

```python
# scale data
import sklearn
from sklearn.preprocessing import MinMaxScaler
scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(-1, 1))
combined_data[["zcr", "spec_cent", "spec_rolloff", "tonnetz"]] = scaler.fit_transform(
                                                                combined_data[["zcr", "spec_cent", "spec_rolloff", "tonnetz"]])
```

```python
sb.pairplot(combined_data, hue="type")
```

#### Problem 4 - Train and Validate a Classifier

You can use any classifier that you prefer, from logistic regression (as seen in STAT 310 and Lecture 7.3) or other classifiers that you might have learned about in DSCI 425.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

data_train, data_test, labels_train, labels_test = train_test_split(
        combined_data[["zcr", "spec_cent", "spec_rolloff", "tonnetz"]],
        combined_data.type,
        test_size=0.3) 
```

```python
LRC = LogisticRegression()
LRC.fit(data_train, labels_train)
```

```python
prediction = LRC.predict(data_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(labels_test, prediction))
```

#### Problem 5 - Visualize Classifier Performance

https://scikit-learn.org/stable/auto_examples/miscellaneous/plot_display_object_visualization.html#sphx-glr-auto-examples-miscellaneous-plot-display-object-visualization-py

```python
# Note: had to update scikit-learn for this because v0.23 is needed for the syntax
# Console: pip install --upgrade scikit-learn
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay


cm = confusion_matrix(labels_test, prediction)
cm_display = ConfusionMatrixDisplay(cm).plot()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay

labels_score = LRC.decision_function(data_test)
fpr, tpr, _ = roc_curve(labels_test, labels_score, pos_label=LRC.classes_[1])
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay

prec, recall, _ = precision_recall_curve(labels_test, labels_score,
                                         pos_label=LRC.classes_[1])
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```
